```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd
```

```python
irisdata = load_iris()
data = pd.DataFrame(data=irisdata.data, columns = irisdata.feature_names )
data['species'] = irisdata.target
data['species'] = data['species'].map({0:'setosa', 1:'versicolor', 2:'verginica'})
data.head()
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x=data['petal length (cm)'], y=data['petal width (cm)'], hue=data['species'], palette = 'bright')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
plt.figure(figsize=(8,8))
for i, col in enumerate(irisdata.feature_names,1):
  plt.subplot(2,2,i)
  sns.histplot(data[col], kde=True, bins=15)
  plt.xlabel(col)
  plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(5,7))
species_count = data['species'].value_counts()
species_name = irisdata.target_names
plt.pie(species_count, labels = species_name, autopct = '%1.1f%%')
plt.title("Frequency count of each flower type")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.pairplot(data, hue='species', palette = 'viridis')
plt.show()
```
